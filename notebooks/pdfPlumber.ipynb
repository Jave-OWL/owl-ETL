```python
import pdfplumber

with pdfplumber.open("unoDeCadaUno/bancolombia_fiducuenta.pdf") as pdf:
    page = pdf.pages[0]
    print(page.extract_text())  # texto "legible" si lo hay
    for word in page.extract_words():
        print(word)  # cada palabra con su bbox
    for char in page.chars[:10]:
        print(char)  # cada caracter con fuente y coordenadas
```

```python
import pdfplumber

with pdfplumber.open("unoDeCadaUno/bancolombia_fiducuenta.pdf") as pdf:
    for page_number, page in enumerate(pdf.pages, start=1):
        print(f"\n--- Página {page_number} ---")

        # Ver texto
        print("Texto:")
        print(page.extract_text() or "No hay texto")

        # Ver imágenes
        if "image" in page.objects:
            print(f"Imágenes: {len(page.objects['image'])}")
        else:
            print("No hay imágenes")

        # Ver vectores
        vectors = page.objects.get("vector", [])
        if vectors:
            print(f"Vectores: {len(vectors)}")
            for obj in vectors[:3]:  # Muestra solo 3 para no saturar
                print(obj)
        else:
            print("No hay vectores en esta página")

        # Ver rectángulos (a veces útiles para detectar celdas o bordes)
        rects = page.objects.get("rect", [])
        if rects:
            print(f"Rectángulos: {len(rects)}")
            for obj in rects[:3]:
                print(obj)
        else:
            print("No hay rectángulos en esta página")
```
